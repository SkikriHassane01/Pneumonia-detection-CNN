<code>What is Pneumonia?</code>
<p>Pneumonia is an inflammatory condition of the lung affecting primarily the small air sacs known as alveoli.Symptoms typically include some combination of productive or dry cough, chest pain, fever and difficulty breathing. The severity of the condition is variable. Pneumonia is usually caused by infection with viruses or bacteria and less commonly by other microorganisms, certain medications or conditions such as autoimmune diseases.Risk factors include cystic fibrosis, chronic obstructive pulmonary disease (COPD), asthma, diabetes, heart failure, a history of smoking, a poor ability to cough such as following a stroke and a weak immune system. Diagnosis is often based on symptoms and physical examination. Chest X-ray, blood tests, and culture of the sputum may help confirm the diagnosis.The disease may be classified by where it was acquired, such as community- or hospital-acquired or healthcare-associated pneumonia.</p>
<code>Discription of the dataset</code>
<p>The dataset is organized into 3 folders (train, test, val) and contains subfolders for each image category (Pneumonia/Normal). There are 5,863 X-Ray images (JPEG) and 2 categories (Pneumonia/Normal).

Chest X-ray images (anterior-posterior) were selected from retrospective cohorts of pediatric patients of one to five years old from Guangzhou Women and Children’s Medical Center, Guangzhou. All chest X-ray imaging was performed as part of patients’ routine clinical care.

For the analysis of chest x-ray images, all chest radiographs were initially screened for quality control by removing all low quality or unreadable scans. The diagnoses for the images were then graded by two expert physicians before being cleared for training the AI system. In order to account for any grading errors, the evaluation set was also checked by a third expert.
 </p>

    🔻🌟 Loading the dataset 🌟🔻</h1>

```python
path_train = '/kaggle/input/chest-xray-pneumonia/chest_xray/train'
path_test = '/kaggle/input/chest-xray-pneumonia/chest_xray/test'
path_val = '/kaggle/input/chest-xray-pneumonia/chest_xray/val'

number_of_images_train = {}
number_of_images_test = {}
number_of_images_val = {}

def number_img_in_dir(path, dic_returned):
    for dir in os.listdir(path):
        dic_returned[dir] = len(os.listdir(os.path.join(path,dir)))
    return dic_returned
        
number_of_images_train = number_img_in_dir(path_train, number_of_images_train)
number_of_images_test = number_img_in_dir(path_test, number_of_images_test)
number_of_images_val = number_img_in_dir(path_val, number_of_images_val)

print("training examples: ",number_of_images_train.items())
print("testing examples: ",number_of_images_test.items())
print("validation examples: ",number_of_images_val.items())
```

### **the training data is imbalanced since there are significantly more images of PNEUMONIA (3875) compared to NORMAL (1341). This imbalance can affect the performance of your model, making it biased towards predicting the majority class (PNEUMONIA). we will use data augmentation to handel this problem**

```python
image_size = 150
labels = ['PNEUMONIA','NORMAL']

# the data list format is like this data= numpy_arry( [data_array, class_number{O or 1}] )

def get_data(path):
    data_images = []
    data_labels = []
    for label in labels:
        full_path = os.path.join(path, label)
        class_number = labels.index(label)
        for img in os.listdir(full_path):
            try:
                img_array = cv2.imread(os.path.join(full_path, img), cv2.IMREAD_GRAYSCALE)
                resized_array = cv2.resize(img_array, (image_size, image_size))  # Resize images to 150x150
                data_images.append(resized_array)
                data_labels.append(class_number)
            except Exception as e:
                print(f"Error processing image {img}: {e}")
    return np.array(data_images), np.array(data_labels)


train_images, train_labels = get_data(path_train)
test_images, test_labels = get_data(path_test)
val_images, val_labels = get_data(path_val)

# Check the shapes of the datasets
print("Training data shape:", train_images.shape, train_labels.shape)
print("Testing data shape:", test_images.shape, test_labels.shape)
print("Validation data shape:", val_images.shape, val_labels.shape)
```

```python
# Normalize the data
train_images = np.array(train_images) / 255
test_images = np.array(test_images) / 255
val_images = np.array(val_images) / 255
```

```python
# resize the dataset with 1 chanel
# reshape(num_sampls, height, width, chanel)
def reshape_df(images):
    images = images.reshape(-1, image_size, image_size, 1)
    images = np.array(images)
    return images
    
    
train_images = reshape_df(train_images)
test_images = reshape_df(test_images)
val_images = reshape_df(val_images)
```

<code>What is Data Augmentation</code>
<p>is a technique used to prevent overfitting and handling the imbalance in a dataset in machine learning by artificially increasing the size of a training dataset. This involves applying small modifications to the existing data, which helps the model learn to recognize variations in the data while keeping the labels the same. Common transformation methods include converting images to grayscale, flipping them horizontally or vertically, cropping them randomly, adjusting colors, and rotating them. By using these methods, you can significantly increase the number of training examples, making your model more robust and improving its performance.</p>
</div>


```python
new_images = ImageDataGenerator(
            featurewise_std_normalization = False,
            samplewise_std_normalization = False,
            samplewise_center = False,
            featurewise_center = False,
            zca_whitening = False,
            rotation_range = 30,
            zoom_range = 0.2,
            width_shift_range = 0.1,
            height_shift_range = 0.1,
            horizontal_flip = True,
            vertical_flip = False )

new_images.fit(train_images)
```

    🔻  Building CNN Model 🚊🔻</h1>

```python
model = Sequential([
    # First convolutional layer
    Conv2D(32, (3, 3), strides=1, padding='same', activation='relu', input_shape=(150, 150, 1)),
    BatchNormalization(),
    MaxPooling2D((2, 2), strides=2, padding='same'),

    # Second convolutional layer
    Conv2D(64, (3, 3), strides=1, padding='same', activation='relu'),
    Dropout(0.1),
    BatchNormalization(),
    MaxPooling2D((2, 2), strides=2, padding='same'),

    # Third convolutional layer
    Conv2D(64, (3, 3), strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2), strides=2, padding='same'),

    # Fourth convolutional layer
    Conv2D(128, (3, 3), strides=1, padding='same', activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    MaxPooling2D((2, 2), strides=2, padding='same'),

    # Fifth convolutional layer
    Conv2D(256, (3, 3), strides=1, padding='same', activation='relu'),
    Dropout(0.3),
    BatchNormalization(),
    MaxPooling2D((2, 2), strides=2, padding='same'),

    # Flatten the output of the convolutional layers
    Flatten(),

    # Fully connected layer
    Dense(128, activation='relu'),
    Dropout(0.4),

    # Output layer with sigmoid activation for binary classification
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

# Model summary
model.summary()
```

    🔻🌟 Training  🌟🔻</h1>
    
`ReduceLROnMultipleAccuracies Callback`: This TensorFlow callback reduces the learning rate when validation accuracy exceeds a set threshold. It's useful for fine-tuning and preventing overfitting by making smaller updates to the model's weights as it learns.
    

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

```python
learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy',
                                            patience = 2,
                                            verbose=1,
                                            factor=0.3,
                                            min_lr=0.000001)


history = model.fit(
    new_images.flow(train_images,
                    train_labels,
                    batch_size = 32),
    epochs = 40,
    validation_data = new_images.flow(test_images,
                                      test_labels),
    callbacks = [early_stopping ,learning_rate_reduction]
)
```

```python
# save the model 
model.save('model.keras')

# Loading saved model
model = load_model('model.keras')

print("Loss of the model is - " , model.evaluate(test_images,test_labels)[0])
print("Accuracy of the model is - " , model.evaluate(test_images,test_labels)[1]*100 , "%")
```

```python
# Save the history
with open('training_history.pkl', 'wb') as file:
    pickle.dump(history.history, file)
    
    
# Load the training history
with open('training_history.pkl', 'rb') as file:
    history = pickle.load(file)

# Extract the data
acc = history['accuracy']
val_acc = history['val_accuracy']
loss = history['loss']
val_loss = history['val_loss']
epochs = range(len(acc))

# Plot accuracy
fig = plt.figure(figsize=(13,7))
plt.title("training vs testing acuuracy")
plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.legend(loc='upper left')
plt.show()

# Plot loss
fig = plt.figure(figsize=(13,7))
plt.title("training vs testing loss")
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.legend(loc='upper left')
plt.show()
```

```python
# predict the val images
predictions_prob = model.predict(val_images)
predictions = (predictions_prob > 0.5).astype(int).reshape(-1)

# print the first 15 
predictions[:15]
```

```python
# predict the test images
predictions_prob = model.predict(test_images)
predictions = (predictions_prob > 0.5).astype(int).reshape(-1)
```

```python
print(classification_report(test_labels, predictions, target_names = ['Pneumonia (Class 0)','Normal (Class 1)']))
```

```python
cm = confusion_matrix(test_labels,predictions)
cm = pd.DataFrame(cm , index = ['0','1'] , columns = ['0','1'])
plt.figure(figsize = (10,10))
plt.title("Confusion matrix")
sns.heatmap(cm,cmap= "Blues", linecolor = 'black' , linewidth = 1 , annot = True, fmt='',xticklabels = labels,yticklabels = labels)
plt.show()
```
